```python
import pandas as pd

data = pd.read_csv("finds_electricity_small.csv")
print(data)
```

```python
# Find-S Algorithm

def find_s(data):
    hypothesis = ["Ø"] * (len(data.columns) - 1)

    for index, row in data.iterrows():
        if row["Class"] == "Yes":
            for i in range(len(hypothesis)):
                if hypothesis[i] == "Ø":
                    hypothesis[i] = row[i]
                elif hypothesis[i] != row[i]:
                    hypothesis[i] = "?"
    return hypothesis

h = find_s(data)

print("Final Hypothesis:", h)
```
